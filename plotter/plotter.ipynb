```python
%matplotlib inline
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'ieee'])
plt.rcParams["text.usetex"] = False
plt.rcParams['figure.figsize'] = 6, 2
```

```python
from_index, to_index = 4000, 4500
```

```python
def plots(name, client_ml_pred, client_ml_true, ml_pred, fl_pred):
    plt.figure(figsize=(6, 2))
    plt.plot(client_ml_true, label='Ground Truth', linewidth=0.4, linestyle='-')
    
    plt.plot(client_ml_pred, label='Client ML', linewidth=0.3, linestyle='--')
    
    plt.plot(ml_pred, label='Central ML', linestyle='--', linewidth=0.3)

    plt.plot(fl_pred, label='FL', linestyle='-', linewidth=0.4)

    plt.title(f'{name}', loc='left', y=1.1, fontsize=11)
    plt.xlabel('Timestamp')
    plt.ylabel('Torque')
    plt.yticks([])

    plt.legend(
        loc='upper right', ncol=4,
        bbox_to_anchor=(0, 1, 1, 0.15),
    )
    plt.show()
```

```python
client_ml_true = pd.read_csv('EPS1/y_true.csv', header=None).to_numpy()
client_ml_pred = pd.read_csv('EPS1/y_pred.csv', header=None).to_numpy()

client_ml_true = client_ml_true[from_index:to_index, 1]
client_ml_pred = client_ml_pred[from_index:to_index, 1]

fl_pred = pd.read_csv('FL1/y_pred.csv', header=None).to_numpy()
fl_pred = fl_pred[from_index:to_index, 1]

ml_pred = pd.read_csv('ML1/y_pred.csv', header=None).to_numpy()
ml_pred = ml_pred[from_index:to_index, 1]
```

```python
plots('Vehicle 1', client_ml_pred, client_ml_true, ml_pred, fl_pred)
```

```python
# RMSE Losses 

from sklearn.metrics import mean_squared_error

print("Client ML RMSE Loss: " + str(mean_squared_error(client_ml_true, client_ml_pred, squared=False)))
print("Central ML RMSE Loss: " + str(mean_squared_error(client_ml_true, ml_pred, squared=False)))
print("FL RMSE Loss: " + str(mean_squared_error(client_ml_true, fl_pred, squared=False)))
```
